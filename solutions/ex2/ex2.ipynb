```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report,confusion_matrix#这个包是评价报告
```

### Data Visualization & Data Preparation

```python
data1 = pd.read_csv('/Users/hsunku/Desktop/github/MachineLearning_AndrewNg/exercises/machine-learning-ex2/ex2/ex2data1.txt',
                     header = None,
                     names = ['Exam_1_score','Exam_2_score','Admitted'])
plt.figure(figsize=(10,9))
sns.scatterplot(x=data1.Exam_1_score,y=data1.Exam_2_score,hue=data1.Admitted)
plt.show()
data1.head()
```

```python
def get_X(df):
    ones = pd.DataFrame({'ones':np.ones(len(df))}) # m by 1 matrix
    data = pd.concat([ones,df],axis=1) #添加intercept
    return np.matrix(data.iloc[:, :-1])#.as_matrix()

def get_y(df):
    return np.array(df.iloc[:,-1])

def normalized_feature(df):
    return df.apply(lambda column: (column-column.mean())/column.std())
```

### Sigmoid function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(np.arange(-10, 10, step=0.01),
        sigmoid(np.arange(-10, 10, step=0.01)))
ax.set_ylim((-0.1,1.1))
ax.set_xlabel('z', fontsize=18)
ax.set_ylabel('g(z)', fontsize=18)
ax.set_title('sigmoid function', fontsize=18)
plt.show()
```

### Cost function

```python
def cost(theta,X,y):
    return np.mean((-y*np.log(sigmoid(X.dot(theta))).T- (1-y)*np.log(1-sigmoid(X.dot(theta))).T)/X.shape[0])
```

```python
X = get_X(data1)
y = get_y(data1)
theta = np.zeros(X.shape[1])

cost(theta, X, y)
```

### gradient descent

```python
def gradient(theta, X, y):
#     '''just 1 batch gradient'''
    output = ((1 / len(X)) * X.T @ (sigmoid(X @ theta) - y).T)
    new_output = []
    for i in range(output.shape[0]):
        new_output.append(output[i,0])
    return new_output
```

```python
gradient(theta, X, y)
```

### Finding the best parameters & Validation

```python
import scipy.optimize as opt
res = opt.minimize(fun=cost, x0=theta, args=(X, y), method='Newton-CG', jac=gradient)
print('best parameters:',res.x)
print('cost:',res.fun)
def predict(x,theta,threshold=0.5):
    prob = sigmoid(x@theta)
    return (prob>=threshold).astype(int)
accuracy=np.mean(predict(X,res.x) == y)*100
print('Accuracy: ', accuracy,'%')
plt.figure(figsize=(12,9))
sns.scatterplot(x=data1.Exam_1_score,y=data1.Exam_2_score,hue=data1.Admitted)
sns.lineplot(x=data1.Exam_1_score,y=(res.x[0]+res.x[1]*data1.Exam_1_score)/-res.x[2],color='r')
plt.show()
```

# Feature Mapping

```python
def feature_mapping(x, y, power, as_ndarray=False):
#     """return mapped features as ndarray or dataframe"""
    # data = {}
    # # inclusive
    # for i in np.arange(power + 1):
    #     for p in np.arange(i + 1):
    #         data["f{}{}".format(i - p, p)] = np.power(x, i - p) * np.power(y, p)

    data = {"f{}{}".format(i - p, p): np.power(x, i - p) * np.power(y, p)
                for i in np.arange(power + 1)
                for p in np.arange(i + 1)
            }

    if as_ndarray:
        return np.matrix(pd.DataFrame(data))
    else:
        return pd.DataFrame(data)
x1=np.array(data2.Microchip_Test_1)
x2=np.array(data2.Microchip_Test_2)
data=feature_mapping(x=x1,y=x2,power=6)
data.head()
```

### Regularized Cost

```python
theta = np.zeros(data.shape[1])
X =feature_mapping(x=x1,y=x2,power=6,as_ndarray=True)
y = get_y(data2)
```

```python
def regularized_cost(theta, X, y, l=1):
#     '''you don't penalize theta_0'''
    theta_j1_to_n = theta[1:]
    regularized_term = (l / (2 * len(X))) * np.power(theta_j1_to_n, 2).sum()

    return cost(theta, X, y) + regularized_term

regularized_cost(theta,X,y,l=1)
```

```python
def regularized_gradient(theta, X, y, l=1):
#     '''still, leave theta_0 alone'''
    theta_j1_to_n = theta[1:]
    regularized_theta = (l / len(X)) * theta_j1_to_n

    # by doing this, no offset is on theta_0
    regularized_term = np.concatenate([np.array([0]), regularized_theta])

    return gradient(theta, X, y) + regularized_term
regularized_gradient(theta, X, y)
```

```python
print('init cost = {}'.format(regularized_cost(theta, X, y)))

res = opt.minimize(fun=regularized_cost, x0=theta, args=(X, y), method='Newton-CG', jac=regularized_gradient)
print('best parameters:',res.x)
print('cost:',res.fun)
print('Accuracy: ', np.mean(predict(X,res.x)==y)*100,'%')
print(classification_report(y,predict(X,res.x).T))
```

```python
print(confusion_matrix(predict(X,res.x).T,y))
print(' TN, FN','\n','FP, TP')
```

```python
def find_decision_boundary(density, power, theta, threshhold):
    t1 = np.linspace(-1, 1.5, density)
    t2 = np.linspace(-1, 1.5, density)

    cordinates = [(x, y) for x in t1 for y in t2]
    x_cord, y_cord = zip(*cordinates)
    mapped_cord = feature_mapping(x_cord, y_cord, power) 

    inner_product = np.matrix(mapped_cord) @ theta

    decision = mapped_cord[np.array((np.abs(inner_product)<2*10**-3))[0]][['f01','f10']]

    return decision.f10, decision.f01
# x,y=find_decision_boundary(density=1000, power=6, theta=feature_mapped_logistic_regression(6,1), threshhold=2 * 10** -3)



```
