# 2. Linear Regression with one variable

## 2.1. Visulization

```python
ex1data = pd.read_csv('/Users/hsunku/Desktop/github/MachineLearning_AndrewNg/exercises/machine-learning-ex1/ex1/ex1data1.txt',
                      header=None,
                     names=['Population','Profit'])
X = ex1data['Population']
y = ex1data['Profit']
plt.scatter(X,y,color='r',marker='x')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()
```

# 2.2. Gradient Descent

```python
def computer_cost(X,y,theta):
    return np.sum(np.power((X*theta.T-y),2))/(2*len(X))
# ex1data['ones']=1
ex1data.insert(0,'ones',1)
cols = ex1data.shape[1]
X = np.matrix(ex1data.iloc[:,:cols-1])
y = np.matrix(ex1data.iloc[:,cols-1:cols])
theta = np.matrix(np.array([0,0]))
```

```python
computer_cost(X,y,theta)
```

```python
def gradient_descent(X,y,theta,alpha,iterations):
    temp = np.matrix(np.zeros(theta.shape))
    no_parameters = theta.shape[1]
    cost = np.zeros(iterations)
    count = 0
    
    while count < iterations:
        error = (X * temp.T) - y
        
        for i in range(no_parameters):
            temp[0,i] = temp[0,i] - alpha*np.sum(np.multiply(error,X[:,i]))/len(X)
            
        theta = temp
        cost[count] = computer_cost(X,y,theta)
        count = count + 1
    
    return theta, cost
```

```python
iterations=2000
alpha = 0.01 #learning rate
grad,cost = gradient_descent(X,y,theta,alpha,iterations)
y_predict=X*grad.T
plt.figure(figsize=(10,9))
plt.plot(X[:,1],y_predict)
plt.scatter(ex1data.iloc[:,1],ex1data.iloc[:,2],color='r',marker='x')
plt.xlabel('Population in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()
print(grad)
```

```python
plt.figure(figsize=(10,9))
plt.plot(np.arange(iterations),cost,'r')
plt.xlabel('iterations')
plt.ylabel('Epoch')
plt.title('Error vs. Training Epoch')
plt.show()
```

## 3.2. Gradient Descent

```python
data2_normalized.insert(0,'ones',1)
cols2 = data2_normalized.shape[1]
X2 = np.matrix(data2_normalized.iloc[:,:cols2-1])
y2 = np.matrix(data2_normalized.iloc[:,cols2-1:cols2])
theta2 = np.matrix(np.array([0,0,0]))
```

```python
iterations=1000
alpha = 0.01 #learning rate
g2, cost2 = gradient_descent(X=X2, y=y2, theta=theta2, alpha=alpha, iterations=iterations)
y2_predict=X2*g2.T
```

## 3.3. Normal Equations

```python
def normalEqn(X, y):
    theta = np.linalg.inv(X.T*X)*X.T*y#X.T@XX.T.dot(X)
    return theta
```

```python
normalEqn(X,y)
```

**We should select appropriate approach based on our need:**


|Normal Equation|Gradient Descent|
|----|----|
|No need to choose alpha|Need to choose alpha|
|No need to iterate|Need many iterations|
|O(n3)|O(kn2)|
|slow if n is large|works well when n is large|
